```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train_.csv',usecols=['GarageQual','FireplaceQu','SalePrice'])
```

```python
df.head()
```

```python
df.isnull().mean()*100
```

```python
df['GarageQual'].value_counts().sort_values(ascending=False).plot(kind='bar')
plt.ylabel('No.of Houses')
```

```python
df['GarageQual'].mode()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df[df['GarageQual']=='TA']['SalePrice'].plot(kind='kde',ax=ax) #Distribution of SalePrice for houses whose GarageQual = TA
df[df['GarageQual'].isnull()]['SalePrice'].plot(kind='kde',ax=ax,color='red') #Distribution of SalePrice for houses whose GarageQual = NA

lines,labels = ax.get_legend_handles_labels()
labels=['Houses with TA','Houses with NA']
ax.legend(lines,labels,loc='best')

plt.title('GarageQual')
```

```python
temp = df[df['GarageQual']=='TA']['SalePrice']
```

```python
df['GarageQual'].fillna('TA',inplace=True)
```

```python
df['GarageQual'].value_counts().plot(kind='bar')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

temp.plot(kind='kde',ax=ax)
df[df['GarageQual']=='TA']['SalePrice'].plot(kind='kde',ax=ax,color='red')

lines,labels = ax.get_legend_handles_labels()
labels = ['Original variable','Imputed variable']
ax.legend(lines,labels,loc='best')
```
