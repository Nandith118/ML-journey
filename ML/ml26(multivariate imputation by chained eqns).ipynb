```python
# Iterative imputer / MICE(Multicariate Imputation by Chained Equations)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = np.round(pd.read_csv('50_Startups.csv')[['R&D Spend','Administration','Marketing Spend','Profit']]/10000)
np.random.seed(9)
df = df.sample(5)
df
```

```python
df = df.iloc[:,0:-1]
df
```

```python
df.iloc[1,0] = np.nan
df.iloc[3,1] = np.nan
df.iloc[-1,-1] = np.nan
```

```python
df
```

```python
df0 = pd.DataFrame()

df0['R&D Spend'] = df['R&D Spend'].fillna(df['R&D Spend'].mean())
df0['Administration'] = df['Administration'].fillna(df['Administration'].mean())
df0['Marketing Spend'] = df['Marketing Spend'].fillna(df['Marketing Spend'].mean())
```

```python
#0th iteration
df0
```

```python
df1 = df0.copy()
df1.iloc[1,0]=np.nan
df1
```

```python
X = df1.iloc[[0,2,3,4],1:3]
X
```

```python
y = df.iloc[[0,2,3,4],0]
y
```

```python
lr = LinearRegression()
lr.fit(X,y)
lr.predict(df1.iloc[1,1:].values.reshape(1,2))
```
