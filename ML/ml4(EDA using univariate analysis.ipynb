```python
import pandas as pd
df = pd.read_csv('Titanic.csv')
```

```python
df.head()
```

For Categorical columns

1. Countplot (bar graph)

```python
import seaborn as sns
```

```python
sns.countplot(df,x='Pclass') #countplot gives count of each category
```

```python
df['Pclass'].value_counts().plot(kind='bar') #We can get bar graph like this too 
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.3f') #autopct means  automatic percentage
```

For Numerical Data

1. Histogram
   (Histogram is used to plot graph in ranges)

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Age'],bins=20) #bins is nothing but no.of ranges should be divided
#we can draw using sns also
```

2. Distplot

```python
sns.distplot(df['Age']) #the curve below is KDE kernel density estimation which indicates pdf 
```
