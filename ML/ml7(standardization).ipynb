```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df = df.iloc[:,2:]
```

```python
df.head()
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased',axis=1),df['Purchased'],
                                                    test_size=0.3,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train) #it will fit the scaler to train set, so that scaler will learn parameters

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train
```

```python
X_train_scaled #we given dataframe(X_train) to scaler but it returned numpy array(X_train_scaled)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
X_train_scaled
```

```python
np.round(X_train.describe(),1) #rounding off to 1 digit after decimal
```

```python
np.round(X_train_scaled.describe(),1)
```

Why Scaling is important ?

Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
lr_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred = lr.predict(X_test)
y_scaled_pred = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Before Scaling',accuracy_score(y_test,y_pred))
print('After Scaling' ,accuracy_score(y_test,y_scaled_pred))
```

Using Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt_scaled = DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
dt_scaled.fit(X_train_scaled,y_train)
```

```python
y_pred = dt.predict(X_test)
y_scaled_pred = dt_scaled.predict(X_test_scaled)
```

```python
print('Before Scaling ',accuracy_score(y_test,y_pred))
print('After Scaling ',accuracy_score(y_test,y_scaled_pred))
```
