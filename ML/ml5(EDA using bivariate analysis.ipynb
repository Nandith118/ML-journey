```python
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
iris = sns.load_dataset('iris')
```

1. Scatterplot (Numerical vs Numerical)

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'])
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex']) #hue is used to split data into groups using different colors
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
#hue differentiates with help of  colours
#style differentiates with help of shapes
#size differentiates with help of size
```

2. BarPlot (Numerical vs Categorical)

```python
titanic = pd.read_csv('Titanic.csv')
```

```python
titanic.head()
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'])
```

3. BoxPlot (Numerical vs Categorical)

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Pclass'])
```

4. Distplot (Numerical vs categorical)

```python
sns.distplot(titanic['Age'])
#this is univariate
#pdf of age 
```

```python
sns.distplot(titanic[titanic['Pclass']==3]['Age']) #gives pdf curve of age of people travelling in third class
sns.distplot(titanic[titanic['Pclass']==1]['Age']) #gives pdf curve of age of people travelling in 1st class
```

5. HeatMap (Categorical vs Categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Sex'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Sex'])) #heatmap is useful to find correlation (strong or weak)
```

```python
titanic.groupby('Pclass')['Age'].mean() #creates 3 groups according to class and then find avg of ages in each class
```

```python
(titanic.groupby('Pclass')['Age'].mean()).plot(kind='bar') #to plot bar graph 
```

6. ClusterMap (categorical vs Categorical)

```python
pd.crosstab(titanic['SibSp'],titanic['Parch'])
```

```python
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Parch'])) #clustermap also shows correlation and also helps to know closeness 
```

7. PairPlot 

```python
iris.head()
```

```python
sns.pairplot(iris) #pairplot gives scatterplot between every pair of columns except diagonal (it gives histogram for diagonal pairs)
```

```python
sns.pairplot(iris,hue='species') #pairplot usually used for numerical columns , so we can include species as hue
```

8. LinePlot (Numerical vs Numerical)

```python
flights.head()
```

```python
new = flights.groupby('year')['passengers'].sum().reset_index() 
#when we use groupby on year ,now year column will become index .So we will reset index
```

```python
new.head()
```

```python
sns.lineplot(x=new['year'],y=new['passengers'])
#lineplot mostly used when x axis has time based column
```

```python
x = flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
sns.clustermap(x)
```
