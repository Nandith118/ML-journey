OneHotEncoding using Sklearn

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=42)
```

```python
df.head()
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
#onehotencoder will create a sparse matrix by default , we need to convert into array or we can make sparse as false
```

```python
X_train_new = encoder.fit_transform(X_train[['fuel','owner']]) #we are applying OHE only on two columns
```

```python
X_train_new
```

```python
X_test_new = encoder.transform(X_test[['fuel','owner']])
```

```python
X_test_new
```

```python
np.hstack((X_train[['brand','km_driven']].values,X_train_new))
#hstack means horizontal stack  (It joins array columnwise)
#Here we are joining columns that are not included in OHE and columns created through OHE
```
