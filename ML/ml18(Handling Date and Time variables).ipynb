```python
#Converting into datetime datatype
df1['date'] = pd.to_datetime(df1['date'])
```

```python
df1.info()
```

```python
df1['date_year'] = df1['date'].dt.year                   #extracting year
df1['date_month'] = df1['date'].dt.month                 #extracting month
df1['date_month_name'] = df1['date'].dt.month_name()     #month name
df1['date_day'] = df1['date'].dt.day 
df1['date_dow'] = df1['date'].dt.dayofweek 
df1['date_dow_name'] = df1['date'].dt.day_name()

df1.drop(columns=['product_id','city_id','orders']).sample(5)
```

```python
# is weekend ?
df1['date_is_weekend'] = np.where(df1['date_dow_name'].isin(['Sunday','Saturday']),1,0)
df1.head()
```

```python
#extract quarter
df1['quarter'] = df1['date'].dt.quarter
df1.head()
```

```python
#extract semester
df1['semester'] = np.where(df1['quarter'].isin([1,2]),1,2)
df1.head()
```

```python
#extract time between dates

import datetime
today = datetime.datetime.today()
today
```

```python
today - df1['date']
```

```python
(today-df1['date']).dt.days
```

```python
#weeks passed
np.round((today-df1['date'])/np.timedelta64(1,'W'),0)
```
