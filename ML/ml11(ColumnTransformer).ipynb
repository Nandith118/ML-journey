```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

Using Column Transformer

```python
from sklearn.compose import ColumnTransformer
```

```python
tr = ColumnTransformer(transformers=[('tr1',SimpleImputer(),['fever']),
                                      ('tr2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
                                       ('tr3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])],remainder='passthrough')
```

```python
tr.fit_transform(X_train)
```

```python
tr.fit_transform(X_train).shape
```

```python
tr.transform(X_test)
```
